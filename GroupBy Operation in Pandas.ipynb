## Pandas DataFrame.groupby()

```python
# The result of Groupby operation is not a DataFrame, but dict of DataFrame objects
```

```python
# There are multiple ways to split any DataFrame object into the group which are as follows:

# 1. obj.groupby('key')
# 2. obj.groupby(['key1','key2'])
# 3. obj.groupby(key,axis=1)
```

```python
# 1. obj.groupby('key')
```

```python
obj = df.groupby('Region')
```

```python
# Access data of particular group
```

```python
obj.get_group('West')
```

```python
for region,data in obj:
    print(region)
    print(data)
```

```python
# 2. obj.groupby(['key1','key2'])
```

```python
obj1 = df.groupby(['Region', 'County'])
```

```python
for title,data in obj1:
    print(title)
    print(data)
```

```python
# 3. obj.groupby(key,axis=1)
```

```python
obj2 = df.groupby('Region', axis=1)
```

```python
obj2.get_group('West')
```

```python
for region,data in obj2:
    print(region)
    print(data)
```
